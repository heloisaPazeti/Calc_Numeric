```python
def myFunc(n1, n2):

    t1 = time.time()
    # dois vetores randomicos
    A = np.random.rand(n1)
    B = np.random.rand(n1)
    
    # dois escalares randomicos
    alpha = np.random.rand() * 10
    beta = np.random.rand() * 10

    
    
    # c = alpha*a + beta*b
    c = (alpha * A) + (beta * B)

    tf1 = time.time() - t1

    # matriz randomica e calcula sua enesima potencia
    m = np.random.rand(n2, n2) * 100

    t2 = time.time()
    power1 = np.linalg.matrix_power(m, 2)
    tfa = time.time() - t2

    t2 = time.time()
    power2 = np.linalg.matrix_power(m, 3)
    tfb = time.time() - t2

    t2 = time.time()
    power3 = np.linalg.matrix_power(m, 4)
    tfc = time.time() - t2

    return tf1, tfa, tfb, tfc
```

## Exerc√≠cio 2

```python
def numberSeq(a):

    result = []
    x0 = 0.1

    result.append(x0)
    
    for n in range(5000):
        result.append(a * result[n] * (1-result[n]))

    return result
```

```python
def mediaFunc(list):
    result = 0
    for i in range(len(list)):
        result += list[i]

    return result / len(list)
```

```python
def varFunc(list):
    result = 0
    media = mediaFunc(list)
    for i in range(len(list)):
        result += np.square(list[i] - media)

    return result / (len(list)-1)
```

```python
# Meus Resultados
result = numberSeq(1)

t1 = time.time()
media = mediaFunc(result)
print('Minha Media: ', media, ' -> Tempo Gasto: ',  time.time() - t1 ,'\n')

t1 = time.time()
var = varFunc(result)
print('Minha Varian√ßa: ', var, ' -> Tempo Gasto: ', time.time() - t1 ,'\n')
```

```python
# Fun√ß√µes Prontas

t1 = time.time()
print('Media Numpy: ', np.mean(result), ' -> Tempo Gasto: ',  time.time() - t1)
t1 = time.time()
print('Variancia Numpy: ', np.var(result), ' -> Tempo Gasto: ',  time.time() - t1)
print('\n')

t1 = time.time()
print('Media Statistics: ', statistics.mean(result), ' -> Tempo Gasto: ',  time.time() - t1)
t1 = time.time()
print('Variancia Statistics: ', statistics.variance(result), ' -> Tempo Gasto: ',  time.time() - t1)
```

## Exerc√≠cio 4

```python
# x = r * x_{n-1}*(1-x_{n-1})
# https://www.youtube.com/watch?v=BcVLeYV7DGQ
# https://devser.com.br/posts/diagrama-de-bifurcacao-do-mapa-logistico/

R = np.linspace(1, 4, 10**4) # valores p/ a entre 1 e 4

X = [] # parametro a 
Y = [] # valores das sequencias

for r in R:
    

    #x = 0.1
    x = np.random.rand() # valor inicial aleat√≥rio
    for n in range(100): # efeito transiente
        x = r*x*(1-x)

    for n in range(100):
        x = r*x*(1-x)
        X.append(r)
        Y.append(x)

plt.plot(X, Y, ls='', marker=',')
plt.show()
```

```python
# SEM EFEITO TRANSIENTE E COM PARAMETRO FIXO

R = np.linspace(1, 4, 10**4) # valores p/ a entre 1 e 4

X = [] # parametro a 
Y = [] # valores das sequencias

for r in R:
    X.append(r)

    x = 0.1
    
    for n in range(100):
        x = r*x*(1-x)

    Y.append(x)

plt.plot(X, Y, ls='', marker=',')
plt.show()
```

## Exerc√≠cio 5

```python
def addLine(node1, node2, nodes, m):
    if(nodes.count(node1)!=0 and nodes.count(node2)!=0):
        m[nodes.index(node1), nodes.index(node2)] = 1
        m[nodes.index(node2), nodes.index(node1)] = 1
        
    return m
```

```python
def deleteLine(node1, node2, nodes, m):
    if(nodes.count(node1)!=0 and nodes.count(node2)!=0):
        m[nodes.index(node1), nodes.index(node2)] = 0
        m[nodes.index(node2), nodes.index(node1)] = 0
        linha1 = m[nodes.index(node1), :]
        linha2 = m[nodes.index(node2), :]

        if(np.count_nonzero(linha1 == 1) == 0):
            m = deleteNode(node1, nodes, m)
        elif(np.count_nonzero(linha2 == 1) == 0):
            m = deleteNode(node2, nodes, m)
            
    return m
```

```python
def deleteNode(node, nodes, m):
    if(nodes.count(node) != 0):
        m = np.delete(m, nodes.index(node), 0) # deletes row
        m = np.delete(m, nodes.index(node), 1) # deletes column
        nodes = np.delete(nodes, nodes.index(node), 0) # delete label in nodes
        
    return m
```

```python
# MATERIAL
    # SOBRE GRAFOS
        # https://www.youtube.com/watch?v=NNZ7jL1X2mI

    # FUN√á√ïES DE DELE√á√ÉO E ADI√á√ÉO
        # https://www.programiz.com/python-programming/methods/list/index
        # https://www.codespeedy.com/delete-a-row-from-a-numpy-matrix-in-python/
        # https://www.statology.org/numpy-count/
        # https://numpy.org/doc/stable/reference/generated/numpy.count_nonzero.html
        # https://python.igraph.org/en/stable/tutorials/articulation_points.html#sphx-glr-tutorials-articulation-points-py

    # DESENHO DO GRAFO
        # https://stackoverflow.com/questions/28533111/plotting-networkx-graph-with-node-labels-defaulting-to-node-name
        # https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw_networkx.html
        # https://matplotlib.org/stable/gallery/color/named_colors.html


# EXEMPLO: 10 N√ìS => MATRIZ 10x10
n = 10
nos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
labeldict = {}
for i in range(len(nos)):
    labeldict[i] = nos[i]

colors = ['r', 'tab:orange', 'y', 'g', 'b', 'mediumblue', 'violet', 'r', 'g', 'c']

# CRIAR ARRAY DE ZEROS
m = np.zeros(shape=(n,n), dtype=np.float64)

# CRIANDO MATRIZ INICIAL DE EXEMPLO
m = addLine('A', 'B', nos, m)
m = addLine('B', 'C', nos, m)
m = addLine('B', 'D', nos, m)
m = addLine('C', 'E', nos, m)
m = addLine('C', 'G', nos, m)
m = addLine('D', 'E', nos, m)
m = addLine('G', 'F', nos, m)
m = addLine('G', 'H', nos, m)
m = addLine('F', 'H', nos, m)
m = addLine('H', 'I', nos, m)
m = addLine('I', 'J', nos, m)
print('MATRIZ INICIAL:\n',m, '\n')
print('REPRESENTA√á√ÉO EM GRAFO:\n')

G = nx.from_numpy_array(m)
nx.draw(G, labels=labeldict, with_labels  = True, node_color=colors, node_size=500)

```

```python
# REMOVENDO LIGA√á√ÉO
m = deleteLine('B', 'C', nos, m)

labeldict = {}
for i in range(len(nos)-1):
    labeldict[i] = nos[i]


print('MATRIZ APOS DELETE B-C: \n', m, '\n')
print('REPRESENTA√á√ÉO EM GRAFO:\n')
G = nx.from_numpy_array(m)
nx.draw(G, labels=labeldict, with_labels  = True, font_color='y')
```

```python
# REMOVENDO NO
m = deleteNode('J', nos, m)
print('MATRIZ APOS DELETAR J: \n', m, '\n')
print('REPRESENTA√á√ÉO EM GRAFO:\n')


# PRECISA DE NOVO PORQUE FOI REMOVIDO, DEVERIA ESTAR DENTRO DA FUNC DE DELE√á√ÉO
labeldict = {}
for i in range(len(nos)-1):
    labeldict[i] = nos[i]


G = nx.from_numpy_array(m)
nx.draw(G, labels=labeldict, with_labels  = True)
```
