# Aula 2 - Gráficos de Funções

## Gráficos 2D

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xi = 0.0
xf = 2.0*np.pi
N = 1000

x = np.linspace(xi, xf, N)
y = 2.0 + np.sin(2.0*x) + 0.05*np.cos(100.0*x)
```

```python
## PLOTANDO FUNÇÃO

plt.plot(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Meu Gráfico')
plt.show()
```

```python
x = np.linspace(xi, xf, N)
y2 = 2.0 + np.cos(2.0*x) + 0.1*np.cos(20.0*x)

plt.plot(x, y, '-r' , x, y2, 'og')
plt.grid()
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Meu Gráfico 2')
plt.legend(['Função 1', 'Função 2'])
plt.show()
```

## Subplots

```python
y3 = np.exp(-0.5*x)
y4 = np.tanh(y2)


fig, ax = plt.subplots(2,2)
    # fig is figure
    # ax is array -> axes
    # see more: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
    #help(fig) <- try this

ax[0,0].plot(x,y, '-r')
ax[0,0].legend(['Função 1'])

ax[0,1].plot(x,y2, 'g')
ax[0,1].grid()

ax[1,0].plot(x,y3, '_y')
ax[1,0].set_ylabel('y(t)')

ax[1,1].plot(x,y4)
```

```python
# see more about figures: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
# see more about subfigures: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subfigures.html

def example_plot(ax, fontsize=12, hide_labels=False):
    pc = ax.pcolormesh(np.random.randn(30, 30), vmin=-2.5, vmax=2.5)
    if not hide_labels:
        ax.set_xlabel('x-label', fontsize=fontsize)
        ax.set_ylabel('y-label', fontsize=fontsize)
        ax.set_title('Title', fontsize=fontsize)
    return pc

np.random.seed(19680808)
# gridspec inside gridspec
fig = plt.figure(layout='constrained', figsize=(10, 4))
subfigs = fig.subfigures(1, 2, wspace=0.07)

axsLeft = subfigs[0].subplots(1, 2, sharey=True)
subfigs[0].set_facecolor('0.75')
for ax in axsLeft:
    pc = example_plot(ax)
subfigs[0].suptitle('Left plots', fontsize='x-large')
subfigs[0].colorbar(pc, shrink=0.6, ax=axsLeft, location='bottom')

axsRight = subfigs[1].subplots(3, 1, sharex=True)
for nn, ax in enumerate(axsRight):
    pc = example_plot(ax, hide_labels=True)
    if nn == 2:
        ax.set_xlabel('xlabel')
    if nn == 1:
        ax.set_ylabel('ylabel')

subfigs[1].set_facecolor('0.85')
subfigs[1].colorbar(pc, shrink=0.6, ax=axsRight)
subfigs[1].suptitle('Right plots', fontsize='x-large')

fig.suptitle('Figure suptitle', fontsize='xx-large')

plt.show()
```

## Gráficos 3D

```python
from mpl_toolkits.mplot3d import Axes3D

def myFunc(x,y):
    return (1-(x**2+y**3))*np.exp(-(x**2+y**2)/2)

x = np.arange(-3.0, 3.0, 0.1)
y = np.arange(-3.0, 3.0, 0.1)
x,y = np.meshgrid(x, y)
z = myFunc(x, y)

fig = plt.figure()
ax = plt.axes(projection="3d", facecolor='w')
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='viridis', edgecolor=None)
ax.set(xlabel ="$x$", ylabel="$y$", zlabel="$f(x,y)$")

m = plt.cm.ScalarMappable(cmap='viridis') #cmap must be the same on ax.plot_surface
m.set_array(z)
plt.colorbar(m, ax=ax)
plt.show()
```

## Histograma

```python
N = 10000
n = 3

somas = []

for i in range(N):
    x = np.random.rand(n)
    s = np.sum(x)
    somas.append(s)

plt.hist(somas, bins=50, histtype="stepfilled")

#see more about hist: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
```

```python
counts, bins = np.histogram(somas)
plt.stairs(counts, bins)
#plt.hist(bins[:-1], bins, weights=counts)
```
